```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv(r"C:\Users\Lovely Upadhyay\Desktop\householdtask3.csv")
```

```python
display(data.head(10))
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])

#adding title
plt.title('Scatter plot')

#setting x and y label
plt.xlabel('year')
plt.ylabel('own')

#showing result
plt.show()
```

```python
#line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#Adding title
plt.title('Line Chart')

#setting x and y label
plt.xlabel('year')
plt.ylabel('own')

#showing result
plt.show()
```

```python
#bar plot with year against own
plt.bar(data['year'],data['own'])

#Adding title
plt.title('Bar Plot')

#setting x and y label
plt.xlabel('year')
plt.ylabel('own')

#showing result
plt.show()
```

```python
#histogram
plt.hist(data['income'])

plt.title('Histogram')

plt.show()
```
